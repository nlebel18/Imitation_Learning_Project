```python
import numpy as np
import matplotlib.pyplot as plt
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy

from imitation.algorithms.adversarial.airl import AIRL
from imitation.data import rollout
from imitation.data.wrappers import RolloutInfoWrapper
from imitation.rewards.reward_nets import BasicShapedRewardNet
from imitation.util.networks import RunningNorm
from imitation.util.util import make_vec_env

# Configuration des hyperparamètres
SEED = 42
ENV_NAME = "LunarLanderContinuous-v2"  # Environnement choisi
ROLLOUT_EPISODES = 30
TRAIN_STEPS = 20000
EVAL_INTERVAL = 5000 
EVAL_EPISODES = 10

# Fonction pour tracer les récompenses moyennes 
def plot_rewards_dagger_style(rewards, steps, eval_intervals):
    x = np.arange(0, steps + 1, eval_intervals)
    plt.figure(figsize=(10, 6))
    plt.plot(x[:len(rewards)], rewards, marker='o')
    plt.title("\u00c9volution de la r\u00e9compense moyenne pendant l'entra\u00eenement AIRL")
    plt.xlabel("\u00c9tapes d'entra\u00eenement")
    plt.ylabel("R\u00e9compense moyenne")
    plt.grid()
    plt.show()

# Étape 1 : Entraînement du modèle expert 
print("Entra\u00eenement de l'expert PPO...")
env = gym.make(ENV_NAME)
expert = PPO("MlpPolicy", env, verbose=1, seed=SEED)
expert.learn(total_timesteps=50000)  # Réduction des étapes pour tester rapidement
expert.save("ppo-CartPole-v1")

# Évaluation de l'expert
mean_reward, std_reward = evaluate_policy(expert, env, n_eval_episodes=10)
print(f"R\u00e9compense moyenne de l'expert : {mean_reward} +/- {std_reward}")

# Étape 2 : Générer des trajectoires expertes
print("G\u00e9n\u00e9ration de trajectoires expertes...")
venv = make_vec_env(
    ENV_NAME,
    rng=np.random.default_rng(SEED),
    n_envs=1,
    post_wrappers=[lambda env, _: RolloutInfoWrapper(env)],
)
rollouts = rollout.rollout(
    expert,
    venv,
    rollout.make_sample_until(min_episodes=ROLLOUT_EPISODES),
    rng=np.random.default_rng(SEED),
)

# Étape 3 : Initialiser l'apprenant
print("Initialisation de l'apprenant PPO...")
learner = PPO(
    env=venv,
    policy="MlpPolicy",
    batch_size=512,
    ent_coef=0.0,
    learning_rate=3e-4,
    gamma=0.99,
    clip_range=0.2,
    vf_coef=0.5,
    n_epochs=10,
    seed=SEED,
)

# Étape 4 : Initialiser le réseau de récompenses pour AIRL
reward_net = BasicShapedRewardNet(
    observation_space=venv.observation_space,
    action_space=venv.action_space,
    normalize_input_layer=RunningNorm,
)

# Étape 5 : Initialiser AIRL
print("Initialisation d'AIRL...")
airl_trainer = AIRL(
    demonstrations=rollouts,
    demo_batch_size=2048,
    gen_replay_buffer_capacity=512,
    n_disc_updates_per_round=16,
    venv=venv,
    gen_algo=learner,
    reward_net=reward_net,
    allow_variable_horizon=True,
)

# Étape 6 : Suivi des récompenses pendant l'entraînement
print("D\u00e9but de l'entra\u00eenement AIRL...")
rewards_over_time = []

for step in range(0, TRAIN_STEPS + 1, EVAL_INTERVAL):
    airl_trainer.train(EVAL_INTERVAL)
    rewards, _ = evaluate_policy(
        learner, venv, n_eval_episodes=EVAL_EPISODES, return_episode_rewards=True
    )
    mean_reward = np.mean(rewards)
    print(f"\u00c9tape {step}/{TRAIN_STEPS} - R\u00e9compense moyenne : {mean_reward:.2f}")
    rewards_over_time.append(mean_reward)

# Étape 7 : Tracer les récompenses
plot_rewards_dagger_style(rewards_over_time, TRAIN_STEPS, EVAL_INTERVAL)
```
